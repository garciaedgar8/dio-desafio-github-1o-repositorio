```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
#leitura do arquivo
df = pd.read_excel("/content/drive/MyDrive/Total_de_Emb.xlsx")
```

```python
# obs: DN: Distrito Naval, OM: Organização Militar
df.head()
```

```python
df.sample(10)
```

```python
#consultando linhas com valores faltantes
df.isnull().sum()
```

```python
#quantidade de embarcações por estado
df.groupby("UF").sum()
```

```python
#Organização militar com maior número de embarcações cadastradas
df["OM"].max()
```

```python
# 10 maiores cadastros por tipo de embarcação e OM
df.nlargest(10, "TOTAL")
```

```python
# OM com menor número de embarcações cadastradas
df["OM"].min()
```

```python
# quantidade de tipos de embarcações cadastradas
df.groupby("TIPO").sum()
```

VISUALIZAÇÃO DOS DADOS

```python
#Gráfico dos estados com mais embarcações cadastradas
df.groupby("UF").sum().plot.bar(title='Total de embarcações cadastradas por UF', color="green")
plt.xlabel("Estados")
plt.ylabel("Nº embarcações")
```
